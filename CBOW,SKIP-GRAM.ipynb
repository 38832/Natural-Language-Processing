```python
from gensim.models import Word2Vec
```

```python
# Sample corpus
corpus = [
    "natural language processing is fun",
    "machine learning enables nlp applications",
    "deep learning enhances nlp models",
    "text preprocessing is essential for nlp"
]
```

```python
# Tokenize the sentences
tokenized_corpus = [sentence.split() for sentence in corpus]
```

```python
# Train the CBOW model
cbow_model = Word2Vec(sentences=tokenized_corpus, vector_size=50, window=3, min_count=1, sg=0)
```

```python
# Save and use the CBOW model
cbow_model.save("cbow_model.model")
```

```python
# Example: Find similar words
similar_words = cbow_model.wv.most_similar("nlp", topn=5)
print("Words similar to 'nlp':", similar_words)
```
